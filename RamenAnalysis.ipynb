## Introduction
This dataset holds several ratings for ramen products and records the brand, name, style, country, and average stars of each product. This dataset can be found at https://www.kaggle.com/residentmario/ramen-ratings. In this analysis, we will be exploring which brand has the highest average ramen rating, which country has the highest average ramen rating, and whcih ramen style is the most common.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("ramen-ratings.csv")
df
```

```python
df.describe()
```

```python
list(df.columns)
```

```python
list(df.index)
```

### Highest Rated Ramen Style

```python
df['Stars'] = pd.to_numeric(df['Stars'], errors = 'coerce')
stats = pd.pivot_table(data=df, 
                        values =['Stars'], 
                        index=['Style'], 
                        aggfunc='mean')
print(stats)
stats.plot()
plt.title("Ramen Rating by Type")
plt.ylabel("Average Rating")
```

### Highest Rated Ramen Brands

```python
df['Stars'] = pd.to_numeric(df['Stars'], errors = 'coerce')
branddf = df.groupby(['Brand']).mean()
branddf = branddf.sort_values('Stars', ascending = False)
print(branddf.loc[branddf['Stars'] == 5.0] )
```

### Countries with the Highest Rated Ramen Products

```python
df['Stars'] = pd.to_numeric(df['Stars'], errors = 'coerce')
countrydf = df.groupby(['Country']).mean()
countrydf = countrydf.sort_values('Stars', ascending = False)
print(countrydf['Stars'])
```

### Most Common Ramen Styles

```python
df['Style'].value_counts().plot(kind='bar')
plt.xlabel("Style")
plt.ylabel("Count")
plt.title("Most Comomon Ramen Styles")
```

### Results
From our analysis of the data set, we found that there are actually many brands that average five star ratings on their products. Some examples are Kimura, Oree Garden, and The Ramen Rater Select. We also found that the countries that the country that produces the highest rated ramen products on average is Brazil. Finally, we found that the most common style of ramen products is pack ramen.
